Определим неполные данные другим способом: по строкам

```python
for col in df.columns:
  missing = df[col].isnull()
  num_missing = np.sum(missing)
  if num_missing > 0:
    print(f'Created missing indicator for: {col}. {num_missing} missing values')
    df[f'{col}_ismissing'] = missing
is_missing_cols = [col for col in df.columns if '_ismissing' in col]
df['num_missing'] = df[is_missing_cols].sum(axis=1)
```

```python
df.sort_values(by='num_missing', ascending=False)
```

Мы добавили колонку с количеством отсутствующих значений в каждой строке.<br>
Для учебных целей удалим строки с количеством отсутствующих значений > 2 <br>
(На практике мы бы взяли этот показатель равным например 13-ти - исключая строки, состоящие преимущественно из пустых значений)

```python
line_with_missing_values = df[df['num_missing']>2]
df_cleaned_rows = df.drop(line_with_missing_values.index)
```

```python
def clean_up_text(array):
  to_lowercase = list(map(lambda x: x.lower(), array))
  transliterate = list(map(lambda x: unidecode(x), to_lowercase))
  return transliterate
```

Написать функцию, которая добавит колонку с разбиением возраста по группам:<br> до 20, от 20 до 30, от 30 до
36 и старше 36. <br>Посчитать количество футболистов в каждой категории.

```python
def add_age_group(df, ages, age_column_name='Age'):
  def arrange_age_groups():
    ages.sort()
    return ages

  def get_age_group(age, age_groups=arrange_age_groups()):
    for idx in range(len(age_groups)):
      if int(age) < age_groups[idx]:
        return f'< {age_groups[idx]}' if idx == 0 else f'{age_groups[idx - 1]}-{age_groups[idx]}'
    return f'> {age_groups[idx]}'



  df['Age_group'] = list(map(get_age_group, df[age_column_name]))
  return df




df_with_age_groups = add_age_group(df, [20, 30, 36])
```

```python
print(f"Количество футболистов в каждой группе:\n {df_with_age_groups.groupby('Age_group')['ID'].count()}")
```
