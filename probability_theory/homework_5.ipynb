## Задача 1. Когда используется критерий Стьюдента, а когда Z –критерий? ![student.jpg](![Frenchie.jpg](student.jpg))

Критерий Стьюдента используют в случаях
- когда размер выборки мал (обычно до 30)
- когда распределение выборки неизвестно: не известно, соответствует ли распределение генеральной совокупности нормальному распределению

Z-критерий используют в случаях
- когда размер выборки велик (обычно больше 30)
- известно стандартное отклонение
- известно распределение выборки

## Задача 2.
Проведите тест гипотезы.<br>
Утверждается, что шарики для подшипников, изготовленные автоматическим станком,<br> имеют средний диаметр 17 мм. <br>Используя односторонний критерий с α=0,05, <br>проверить эту гипотезу, <br>если в выборке из n=100 шариков средний диаметр оказался равным 17.5 мм, <br>а дисперсия известна и равна 4 кв. мм.

## Решение
Размер выборки > 30 - используем критерий односторонний Z-тест, так как дисперсия известна и размер выборки достаточно велик

Нулевая гипотеза (H0): средний диаметр шариков для подшипников равен 17 мм. Альтернативная гипотеза (H1) утверждает, что средний диаметр больше 17 мм

Формула Z-статистики:

$$Z=\frac{{\over X}- \mu}{\frac {\sigma}{\sqrt n}}=\frac{17.5-17}{\frac {\sqrt 4}{\sqrt 100}}=\frac {0.5}{0.2}=2.5$$

 при уровне значимости α=0,05

 Критическое значение Z можно найти в таблице стандартного нормального распределения. Для α=0,05 в одностороннем тесте критическое значение Z составляет примерно 1.645 (ячейки F18, G18 на картинке)

 ![](https://exceltable.com/funkcii-excel/images/funkcii-excel173-5.png)



наше рассчитанное значение Z (2.5) больше критического значения Z (1.645) <br>
Следовательно, мы отвергаем нулевую гипотезу: <br>
Утверждение, что шарики для подшипников, изготовленные автоматическим станком,<br>
имеют средний диаметр 17 мм<br>
не верно.


## Задача 3.<br>
Проведите тест гипотезы. Продавец утверждает, что средний вес пачки печенья составляет 200 г.<br> Из партии извлечена выборка из 10 пачек. Вес каждой пачки составляет: 202, 203, 199, 197, 195, 201, 200, 204, 194, 190.<br> Известно, что их веса распределены нормально. <br>Верно ли утверждение продавца, если учитывать, что доверительная вероятность равна 99%? <br>
(Провести двусторонний тест.)

Решение:<br>
размер выборки невелик (10 пачек), а распределение нормально.<br>
Нулевая гипотеза: Утверждение продавца верно.<br>
Альтернативная гипотеза: Утверждение продавца не соответствует действительности.<br>
Критерий $$\alpha=\frac {0.01}{2}=0.005$$
(делим на два, поскольку проводим двусторонний тест)<br>
Среднее значение весов в выборке
$$\overline X = \frac {202+203+199+197+195+201+200+204+194+190}{10}=$\frac {1985}{10}=198.5$$
Среднее значение генеральной совокупности<br>
$$\mu=200$$

Стандартное отклонение выборки
$$\sigma = \sqrt {\frac {\sum {(x_i-\overline x)^2}}{n-1}}$$
$$\sigma = \sqrt {\frac {(202-198.5)^2+(203-198.5)^2+(199-198.5)^2+(197-198.5)^2+(195-198.5)^2+(201-198.5)^2+(200-198.5)^2+(204-198.5)^2+(194-198.5)^2+(190-198.5)^2}{10-1}}=4.45346$$
Рассчитаем Критерий Стьюдента
$$t_H=\frac{{\overline X}- \mu}{\frac {\sigma_H}{\sqrt n}}=\frac {198.5-200}{\frac {4.45346}{\sqrt 10}} = -1.065108138$$
Теперь сравним расчетное значение с табличным.
Согласно таблице для коэффициента Стьюдента, при доверительной вероятности 0.995
$$1- \frac {1-0.99}{2}=0.995$$
и числе степеней свободы $$k=10-1=9$$
$$t_H=3.250$$
И в обратную сторону (поскольку тест двусторонний)
$$t_H=-3.250$$
Рассчитанное значение коэффициента Стьюдента находится внутри интервала +/- 3,250

Следовательно, среднее значение выборки принадлежит генеральной совокупности и нулевая гипотеза верна.

# Задача 4 <br>
4.Есть ли статистически значимые различия в росте дочерей? <br>Рост матерей 172, 177, 158, 170, 178,175, 164, 160, 169, 165 <br>Рост взрослых дочерей: 173, 175, 162, 174, 175, 168, 155, 170, 160

Решение


```python
import numpy as np
from scipy import stats

moms_array = np.array([172, 177, 158, 170, 178,175, 164, 160, 169, 165])
daughters_array = np.array([173, 175, 162, 174, 175, 168, 155, 170, 160])
t_statistic, p_value = stats.ttest_ind(daughters_array, moms_array)
```

```python
print("t-статистика:", t_statistic,"\np-значение:", p_value)
```

Выводы:
среднее значение группы матерей немного меньше среднего значения дочерей,<br> т.е. дочери в среднем выше<br>
p-значение велико, а значит, велика вероятность получить такие же результаты при условии, что нулевая гипотеза верна.<br>
Следовательно, различия не являются статистически значимыми.
