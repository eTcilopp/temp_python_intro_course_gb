Составьте шаблон дизайна эксперимента для гипотезы, которая набрала больше всего баллов в практическом задании предыдущего урока

# Решение<br>
##Гипотеза <br>

| IF BY | WILL |  BECAUSE|
|-------|------|---------|
|если разместить на сайте положительные <br>отзывы от пользователей | время, проведенное на сайте<br> увеличится минимум на 50% | потому, что посетители будут читать отзывы других пользователей,<br>а это занимает какое-то время.|

## <br> Целевая аудитория <br>
Потенциальные Клиенты. <br>Это могут быть пользователи, которые искали схожие продукты или услуги в интернете, но еще не посетили наш сайт.<br> Это поможет понять, как отзывы влияют на новых посетителей.
# <br> Метрики <br>
Среднее время на сайте (Average Time on Site) - Основная метрика<br>
Показатель отказов (Bounce Rate) - Контр-метрика
# <br> Сегмент <br>
Новые пользователи.<br>
Я предполагаю, что наличие положительных отзывов в первую очередь повлияет на новых  пользователей,<br>которые принимают решение о подписке
# <br> Продолжительность эксперимента <br>
2 недели
# <br> Критерий успеха <br>
Статистически значимое подтверждение альтернативной гипотезы <br>
(предлагаемые изменения оказывают влияние на среднее время на сайте)<br>
Уменьшился или не увеличился Bounce Rate<br>
Общая конверсия на сайте не ухудшилась.
