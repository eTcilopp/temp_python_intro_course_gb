# Шаблон эксперимента 1
## Гипотеза:
Изменение цвета кнопки на оранжевую увеличит конверсию на 5% потому, что, <br>
согласно неизвестно кем проведенным исследованиям оранжевые кнопки имеют <br>
больший click rate
## Что делаем:
Меняем цвет кнопки "Присоединиться" на лэндинг-странице с голубой на оранжевую.
## На каких пользователях тестируем
Тестируем на всех посетителях сайта в пропорции 50/50
## Ключевые метрики для оценки эксперимента
Конверсия в клик на кнопку<br>
Конверсия в скачивание приложения<br>
Конверсия в подписку<br>
Ожидаем, что увеличение составит не менее 5% по каждой из метрик
