Решение.<br>
В качестве продукта выбираю приложение для копи-трейдинда Fabulus Trade<br>
https://fabulustrade.ru/
<br>
<br>
Ключевые метрики: <br>
Количество Сессий (Sessions)<br>
Процент Новых Сессий (Percentage of New Sessions)<br>
Глубина Просмотра (Page Depth)<br>
Время на Сайте (Average Session Duration)<br>
Показатель Отказов (Bounce Rate)<br>
Конверсия (Conversion Rate)<br>
<br><br>
Все эти метрики можно измерить.<br><br>
- Количество сессий - важнейший показатель для онлайн-бизнеса. При правильно поставленном маркетинге количество сессий будет расти.<br>
- Бизнес модель основана на продаже подписок.
Подписавшиеся пользователи не будут ходить на сайт часто: основная работа происходит в приложении<br>нам нужны новые визитеры на сайте, поэтому нужно отслеживать процент новых сессий.<br>
- Глубина просмотра и время на сайте важны для выявления "правильных" клиентов. Можно отсеять посетителей, проведших на сайте совсем мало времени, и сфокусироваться на аудитории, которой наша услуга интересна.<br>
- Показатель отказов позволяет выявить "больные" места на сайте: важен не столько сам процент отказов, сколько распределение этих отказов по страницам. Страницы с бОльшим процентом отказов следует менять.<br>
- Конверсия в подписку - важнейший показатель для бизнеса. Усилия команды должны быть сосредоточены на росте конверсии.
